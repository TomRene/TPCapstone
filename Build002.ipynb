```python
# Load the analysis file for further processing
load='IBMAnalysis1'
stocktable = pd.read_csv('/Users/tompichard/Documents/Personal/NEWPERSONAL/GADS/1.CAPSTONE/DevFiles/DATA/' + load + '.csv')
stocktable = stocktable.rename(columns = {'Unnamed: 0': 'Counter'})
```

```python
#Seed the feature columns (see Write-up)
stockmeta1 = stocktable[(stocktable['Type']=='LLOW')]
stockmeta1['Run1'] = 'NaN'
stockmeta1['Run2'] = 'NaN'
stockmeta1['Rise1'] = 'NaN'
stockmeta1['Rise2'] = 'NaN'
stockmeta1['NLow2'] = 'NaN'
stockmeta1['NLow3'] = 'NaN'
stockmeta1['dslope1'] = 'NaN'
stockmeta1['dslope2'] = 'NaN'
stockmeta1['OULine'] = 'NaN'
stockmeta1['OU'] = 'NaN'
stockmeta1['OUPct'] = 'NaN'
```

```python
#calculate the features and attributes
ctr=0
while ctr < len(stockmeta1):   
    if stockmeta1.at[stockmeta1.index[ctr],'Type3'] == 'CT1':    
        # calculate & store Run1
        run1 = stockmeta1.at[stockmeta1.index[ctr],'Run1'] = stockmeta1.index[ctr+1] - stockmeta1.index[ctr]
        # calculate & store Run2
        run2 = stockmeta1.at[stockmeta1.index[ctr],'Run2'] = stockmeta1.index[ctr+2] - stockmeta1.index[ctr+1]        
        # calculate & store Rise1
        rise1 = stockmeta1.at[stockmeta1.index[ctr],'Rise1'] = stockmeta1.at[stockmeta1.index[ctr+1], 'NLow'] - stockmeta1.at[stockmeta1.index[ctr], 'NLow']
        # calculate & store Rise2        
        rise2 = stockmeta1.at[stockmeta1.index[ctr],'Rise2'] = stockmeta1.at[stockmeta1.index[ctr+2], 'NLow'] - stockmeta1.at[stockmeta1.index[ctr+1], 'NLow']        
        # retrieve & store NLow2
        nlow2 = stockmeta1.at[stockmeta1.index[ctr],'NLow2'] = stockmeta1.at[stockmeta1.index[ctr+1], 'NLow']
        # retrieve & store NLow3
        nlow3 = stockmeta1.at[stockmeta1.index[ctr],'NLow3'] = stockmeta1.at[stockmeta1.index[ctr+2], 'NLow']
        # calculate & store dslope1
        ds1 = stockmeta1.at[stockmeta1.index[ctr],'dslope1'] = \
               ((stockmeta1.at[stockmeta1.index[ctr+1], 'NLow'] - stockmeta1.at[stockmeta1.index[ctr], 'NLow']) / stockmeta1.at[stockmeta1.index[ctr], 'NLow']) / \
               (stockmeta1.index[ctr+1] - stockmeta1.index[ctr])
        # calculate & store dslope2
        ds2 = stockmeta1.at[stockmeta1.index[ctr],'dslope2'] = \
               ((stockmeta1.at[stockmeta1.index[ctr+2], 'NLow'] - stockmeta1.at[stockmeta1.index[ctr+1], 'NLow']) / stockmeta1.at[stockmeta1.index[ctr+1], 'NLow']) / \
               (stockmeta1.index[ctr+2] - stockmeta1.index[ctr+1])
        # calculate projected low3 from slope1 to compare with actual low3 
        OUpt = stockmeta1.at[stockmeta1.index[ctr],'OULine'] = ((ds1 * nlow2) * run2) + nlow2        
        #ComparePoint = ((dslope1 * Low2) * Interval2) + Low2
        # Determine the Over Under
        if nlow3 > OUpt:
            OU = stockmeta1.at[stockmeta1.index[ctr],'OU'] = 'O'
        elif nlow3 < OUpt:
            OU = stockmeta1.at[stockmeta1.index[ctr],'OU'] = 'U'
        elif nlow3 == OUpt:
            OU = stockmeta1.at[stockmeta1.index[ctr],'OU'] = 'Ev'
        # display the actual over under
        OUPercent = stockmeta1.at[stockmeta1.index[ctr],'OUPct'] = (nlow3 - OUpt)/nlow3

        print ctr, stockmeta1.at[stockmeta1.index[ctr], 'Counter'],stockmeta1.at[stockmeta1.index[ctr], 'Type3'], \
        rise1,rise2, stockmeta1.at[stockmeta1.index[ctr], 'NLow']

    ctr = ctr + 1    
```

```python
# Merge the metadata frame with the original dataframe 
stockpredict1 = pd.merge(left=stocktable,right=stockmeta1, how='left')
```

```python
#Seed the performance metrics and attributes
# Entry = buy price; WL1 = W or L for that entry as a result of scenario 1
# PLpct1 = raw % gain or loss resulting from exit in scenario 1
stockpredict = stockpredict1

stockpredict['Entry'] = 'NaN'
stockpredict['PLavg1'] = 'NaN'
stockpredict['PLpct1'] = 'NaN'
stockpredict['WL1'] = 'NaN'
stockpredict['PLavg2'] = 'NaN'
stockpredict['PLpct2'] = 'NaN'
stockpredict['WL2'] = 'NaN'
```

```python
# populate the performance metrics and attributes
ctr2=0
while ctr2 < len(stockpredict):
    if stockpredict.at[stockpredict.index[ctr2],'Type3'] == 'CT1':    
        # assign the CT1 window
        CT1End = stockpredict.at[stockpredict.index[ctr2],'Run1'] + stockpredict.at[stockpredict.index[ctr2],'Run2']
        # determine entry point
        #print stockpredict.at[stockpredict.index[ctr2,'NOpen']        
        Entry = stockpredict.at[stockpredict.index[ctr2+int(CT1End)+1],'NOpen']        
        stockpredict.at[stockpredict.index[ctr2], 'Entry'] = Entry
        #print CT1End, Entry        
        # determine average of week following Entry and the second following
        PLav1 = stockpredict.at[stockpredict.index[ctr2],'PLavg1'] = stockpredict['Adj_Close'][ctr2+CT1End+1:ctr2+CT1End+6].mean()
        PLav2 = stockpredict.at[stockpredict.index[ctr2],'PLavg2'] = stockpredict['Adj_Close'][ctr2+CT1End+6:ctr2+CT1End+11].mean()
        # stockmeta1['NLow'][100:200].mean()        
        # Determine the pct winds for avg1 and avg2
        PLpc1 = stockpredict.at[stockpredict.index[ctr2],'PLpct1'] = (PLav1-Entry)/Entry
        PLpc2 = stockpredict.at[stockpredict.index[ctr2],'PLpct2'] = (PLav2-Entry)/Entry
        if PLpc1 > 0:
            stockpredict.at[stockpredict.index[ctr2],'WL1'] = 'W'
        else:
            stockpredict.at[stockpredict.index[ctr2],'WL1'] = 'L' 
        if PLpc2 > 0:
            stockpredict.at[stockpredict.index[ctr2],'WL2'] = 'W'
        else:
            stockpredict.at[stockpredict.index[ctr2],'WL2'] = 'L'    
        print ctr2, CT1End, Entry, PLav1, PLav2, PLpc1, PLpc2    
    ctr2 = ctr2 + 1
```
